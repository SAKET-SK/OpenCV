```python
path = os.getcwd()
ImageDetected = [f for f in os.listdir(path) if f.endswith('.jpg')]
```

```python
images = ImageDetected

lst = []
for k in images:
    temp = k.split('.')
    first = temp[0]
    lst.append(int(first))

lst.sort()
temp = []
for k in lst:
    temp.append(str(k) + ".jpg")

images = temp
```

```python
images
```

```python
count_car = []
count_person = []
count_motorcycle = []
count_truck = []
#list
for i in range(len(images)):
    a, b, c, d = count(images[i])
    count_car.append(a)
    count_person.append(b)
    count_motorcycle.append(c)
    count_truck.append(d)
```

```python
print(count_car)
print(count_person)
print(count_motorcycle)
print(count_truck)
```
